```python
import pandas as pd

df = pd.read_csv('../data/fbget_original_full_data_2016_and_2017.csv')
df.dropna(subset = ['message'], inplace = True)
print(df.shape)
```

```python
persons = []
organizations = []

for i,d in df[0:100000].iterrows():

    doc = nlp(d.message)

    if len(doc.ents) > 0:
        for ent in doc.ents:
            if ent.label_ == 'PERSON':
                persons.append(str(ent))
            elif ent.label_ == 'ORG':
                organizations.append(str(ent))
                
print(sorted(list(set((persons)))))

persons2= []

for i,d in df[100001:200000].iterrows():

    doc = nlp(d.message)

    if len(doc.ents) > 0:
        for ent in doc.ents:
            if ent.label_ == 'PERSON':
                persons2.append(str(ent))

print(sorted(list(set((persons2)))))

persons3 = []
for i,d in df[200001:300000].iterrows():

    doc = nlp(d.message)

    if len(doc.ents) > 0:
        for ent in doc.ents:
            if ent.label_ == 'PERSON':
                persons3.append(str(ent))
    

print(sorted(list(set((persons3)))))

persons4 = []
for i,d in df[300001:400000].iterrows():

    doc = nlp(d.message)

    if len(doc.ents) > 0:
        for ent in doc.ents:
            if ent.label_ == 'PERSON':
                persons4.append(str(ent))
            
print(sorted(list(set((persons4)))))

persons5 = []
for i,d in df[400001:].iterrows():

    doc = nlp(d.message)

    if len(doc.ents) > 0:
        for ent in doc.ents:
            if ent.label_ == 'PERSON':
                persons5.append(str(ent))
print(sorted(list(set((persons5)))))

```
