```python

import spacy
nlp=spacy.load('en')
import pandas as pd


DATA_PATH="../emile/"
filename="test_colonne_vide_juste.csv"
df=pd.read_csv(DATA_PATH+filename)
#df_tokens=df[["sent_tokens"]]
#print(df_tokens)


persons=[]
organisations=[]
for i,d in df.iterrows():
    doc=nlp(d.sent_tokens)
    if len(doc.ents)>0:
        for ent in doc.ents:
            if ent.label_ == 'PERSON':
                persons.append(str(ent))
            elif ent.label_ == 'ORG':
                organisations.append(str(ent))
print(sorted(list(set(persons))))
print(sorted(list(set(organisations))))
```
