```python
import spacy
nlp=spacy.load('en')
import pandas as pd


DATA_PATH="../emile/"
filename="octobre_corpus.csv"
df=pd.read_csv(DATA_PATH+filename)
#df_tokens=df[["sent_tokens"]]
#print(df_tokens)

persons=[]
organisations=[]
for i,d in df.iterrows():
    doc=nlp(d.message)
    if len(doc.ents)>0:
        for ent in doc.ents:
            if ent.label_ == 'PERSON':
                persons.append(str(ent))
            elif ent.label_ == 'ORG':
                organisations.append(str(ent))
#print(sorted(list(set(persons))))
#print(sorted(list(set(organisations))))

variable=df.to_string(header=True, na_rep='')

variable2=variable.replace("\n", "")
#print(variable2)
#type(variable2)

liste_test=variable2.split(" ")
while '' in liste_test:
    liste_test.remove('')
#print(liste_test)
#print(len(liste_test))

liste_sans_persons=[]
liste_persons=sorted(list(persons))
for element in liste_test:
    if element not in liste_persons:
        liste_sans_persons.append(element)
#print(liste_sans_persons)
#print(len(liste_sans_persons))
dataf = pd.DataFrame({'mots':liste_sans_persons})
print(dataf)
dataf.to_csv("data_oct.csv")
```
